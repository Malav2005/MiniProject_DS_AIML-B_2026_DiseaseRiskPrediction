```python
import pandas as pd
import numpy as np

df = pd.read_excel("../dataset/raw_data/diabetes.csv.xlsx")
```

```python
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df[cols] = df[cols].replace(0, np.nan)
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("../dataset/processed_data/cleaned_data.csv", index=False)
```
